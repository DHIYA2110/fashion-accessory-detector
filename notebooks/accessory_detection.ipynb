```python
import os
import shutil
import cv2

source_folder = '../data/glasses_dataset/sunglasses'
clean_folder = '../data/glasses_dataset/sunglasses_clean'

os.makedirs(clean_folder, exist_ok=True)

count = 0

for file in os.listdir(source_folder):
    src = os.path.join(source_folder, file)
    dst = os.path.join(clean_folder, file)

    # ✅ Copy only if file exists AND can be opened by OpenCV
    if os.path.isfile(src):
        try:
            img = cv2.imread(src)
            if img is not None:
                shutil.copy(src, dst)
                count += 1
        except:
            print(f"Skipped unreadable: {file}")

print(f"✅ Copied {count} clean images to sunglasses_clean folder.")
```

```python
import os
import cv2

folder_path = '../data/glasses_dataset/sunglasses'

broken_files = []

# Check which files exist and can be opened
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    # Double check if the file actually exists
    if not os.path.isfile(file_path):
        print(f"File does not exist: {filename}")
        broken_files.append(filename)
        continue

    # Check if the image can be read
    img = cv2.imread(file_path)
    if img is None:
        print(f"Unreadable image: {filename}")
        broken_files.append(filename)

print(f"Total problematic files: {len(broken_files)}")
```

```python
import shutil

# Create a folder for broken files
os.makedirs('../data/glasses_dataset/broken_files', exist_ok=True)

# Move broken files to a separate folder
for filename in broken_files:
    src_path = os.path.join(folder_path, filename)
    dest_path = '../data/glasses_dataset/broken_files/' + filename

    if os.path.isfile(src_path):
        shutil.move(src_path, dest_path)
        print(f"Moved broken file: {filename}")
```

```python
import os
import shutil
import cv2

source_folder = '../data/glasses_dataset/sunglasses'
clean_folder = '../data/glasses_dataset/sunglasses_clean'

# Step 1: Remove old clean folder if it exists
shutil.rmtree(clean_folder, ignore_errors=True)
os.makedirs(clean_folder, exist_ok=True)

count = 0
skipped = 0

for file in os.listdir(source_folder):
    src = os.path.join(source_folder, file)
    dst = os.path.join(clean_folder, file)

    # ✅ Check: File must physically exist
    if not os.path.isfile(src):
        print(f"Missing file: {file}")
        skipped += 1
        continue

    # ✅ Check: File must be a readable image
    try:
        img = cv2.imread(src)
        if img is None:
            print(f"Unreadable image: {file}")
            skipped += 1
            continue

        # ✅ Copy the clean, existing image
        shutil.copy(src, dst)
        count += 1
    except Exception as e:
        print(f"Error reading file {file}: {e}")
        skipped += 1
        continue

print(f"\n✅ Finished cleaning.")
print(f"Total valid images copied: {count}")
print(f"Total files skipped (missing/unreadable): {skipped}")
```

```python
results = model.predict(
    source='../data/glasses_dataset/sunglasses_clean',
    save=True,
    project='../visuals',
    name='sunglasses_detected_clean'
)

print("✅ Detection completed. Check visuals/sunglasses_detected_clean.")
```
